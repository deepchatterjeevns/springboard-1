```python
# Reading in the API URL
AFX = requests.get('https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?' + API_KEY)
```

```python
1. # Making the Time-span 2017:
TS_AFX = requests.get('https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&' + API_KEY)
```

```python
2. # Converting the JSON object to a python dictionary
json_data = TS_AFX.json()
print(json_data)
```

```python
#We see the column names we want for the dictionary: Date, Open, High, Low, Close, Change 
columns = json_data['dataset']['column_names']
```

```python
# Viewing the data
data = json_data['dataset']['data']
```

```python
# Calculating the highest opening price for the stock prices in the period
open_high = [price[1] for price in data if price[1] != None]
open_high.sort(reverse = True)
print(open_high[0])
```

```python
# Calculating the lowest opening price for the stock price in the period
open_low = [price[1] for price in data if price[1] != None]
open_low.sort(reverse = False)
print(open_low[0])
```

```python
# Finding the highest change in the any one day
change = [price[2] - price[3] for price in data if (price[2] != None) and (price[3] != None)]
change.sort(reverse= True)
print(change[0])
```

```python
# Finding highest change between two days (based on closing)
daily_change =[data[i][4] - data[i-1][4] for i in range(1,len(data)) 
              if (data[i][4] != None) and (data[i-1][4] != None) ]
daily_change.sort(reverse = True)
print(daily_change[0])
```

```python
# Finding average daily trading volume
trading_volume = [vol[6] for vol in data if vol != None]
avg_volume = [sum(trading_volume)/len(trading_volume)]
print(avg_volume)
```
